### 4 Considering only NTAs with at least 5 listed businesses, what is the average total savings and the total jobs created for each NTA?


```python
ve['NTA_count'] = ve.groupby('NTA')['NTA'].transform('count')
ve.head()
```

```python
pd.set_option("display.max_columns", None)
nta5 = ve[ve["NTA_count"] >=5]
nta5 = nta5[["NTA" , "Effective Date" , "NTA_count", "Job created", "Savings from beginning receiving benefits"]]
nta5.rename(columns = {"Effective Date":"Date"  , "Job created":"Jobs", "Savings from beginning receiving benefits":"Savings"}, inplace = True)

nta5.head()
```

```python
nta5["SumJobs"] = nta5.groupby('NTA').Jobs.transform('sum')
nta5["MeanSavings"] = nta5.groupby('NTA').Savings.transform('mean')
nta5.head()
```
